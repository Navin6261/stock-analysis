```python
#importing the libraries

import pandas as pds
import datetime
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#assigning the start and end dates

start_date = datetime(2022,1,3)
end_date = datetime(2022,4,29)
```

```python
#reading the excel file

data = pds.read_excel("D:\internship hw\sample_data(1).xls")
```

```python
df = pds.DataFrame(data)
```

```python
df.head()
```

```python
#converting the object type to datetime

df['date'] = pds.to_datetime(df['date'])
df['date']
```

```python
#extracting time from the datetime frame

df['time'] = df['date'].dt.time
```

```python
#extracting date from the datetime frame

df['date_ext'] = df['date'].dt.date
```

```python
#converting the data frame from datetime to string type

df['time']= df['time'].astype(str)
```

```python
#plotting the graph between open_price, close_price and date

plt.figure(figsize=(20,8))
plt.plot(df.date_ext,df.open, label='open price')
plt.plot(df.date_ext,df.close, label = 'close price',color='red')
plt.legend()
plt.xlim(start_date,end_date)
plt.xticks(rotation=90)
plt.show()
```

```python
#plotting the graph between high_price, low_price and date

plt.figure(figsize=(20,8))
plt.plot(df.date_ext,df.high, label='high price',color='green')
plt.plot(df.date_ext,df.low,label='low price',color='black')
plt.legend()
plt.xlim(start_date,end_date)
plt.xticks(rotation=90)
plt.show()
```

```python
#plotting the graph among open_price, close_price, high_price, low_price and date

plt.figure(figsize=(20,8))
plt.plot(df.date_ext,df.open,label='open price')
plt.plot(df.date_ext,df.close,label='close price')
plt.plot(df.date_ext,df.high,label='high price')
plt.plot(df.date_ext,df.low,label='low price',color='black')
plt.legend()
plt.title('stock prices',color='green',size=30)
plt.xlabel('date',size=20)
plt.ylabel('price',size=20)
```

```python
#volume traded graph
plt.figure(figsize=(20,8))
plt.bar(df.date_ext,df.volume)
plt.show()
```

```python
#volume and open price 
plt.figure(figsize=(20,8))
plt.bar(df.open,df.volume)
plt.title('open price vs Volume traded',size=30 )
plt.xlabel('open price',size = 20)
plt.ylabel('volume traded',size=20)
plt.show()
```
